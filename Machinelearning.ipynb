```python
from PIL import Image
import os

size_300 = (300, 300)

input_folder1 = '/content/dataset_new/test/Closed/'
input_folder2 = '/content/dataset_new/test/Open/'
input_folder3 = '/content/dataset_new/test/no_yawn/'
input_folder4 = '/content/dataset_new/test/yawn/'

output_folder1 = 'dataset_300/test/Closed/'
output_folder2 = 'dataset_300/test/Open/'
output_folder3 = 'dataset_300/test/no_yawn/'
output_folder4 = 'dataset_300/test/yawn/'

for file_object in os.listdir(input_folder1):
  if file_object.endswith('.jpg'):
    file_path = os.path.join(input_folder1, file_object)
    img = Image.open(file_path)
    fn, fext = os.path.splitext(file_object)
    img.thumbnail(size_300)
    output_path = os.path.join(output_folder1, '{}_300{}'.format(fn, fext))
    img.save(output_path)
for file_object in os.listdir(input_folder2):
  if file_object.endswith('.jpg'):
    file_path = os.path.join(input_folder2,file_object)
    img = Image.open(file_path)
    fn,fext = os.path.splitext(file_object)
    img.thumbnail(size_300)
    output_path = os.path.join(output_folder2,'{}_300{}'.format(fn,fext))
    img.save(output_path)
for file_object in os.listdir(input_folder3):
  if file_object.endswith('.jpg'):
    file_path = os.path.join(input_folder3,file_object)
    img = Image.open(file_path)
    fn,fext = os.path.splitext(file_object)
    img.thumbnail(size_300)
    output_path = os.path.join(output_folder3,'{}_300{}'.format(fn,fext))
    img.save(output_path)
for file_object in os.listdir(input_folder4):
  if file_object.endswith('.jpg'):
     file_path = os.path.join(input_folder4,file_object)
     img       = Image.open(file_path)
     fn,fext   = os.path.splitext(file_object)
     img.thumbnail(size_300)
     output_path = os.path.join(output_folder4,'{}_300{}'.format(fn,fext))
     img.save(output_path)



```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
import PIL
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
data_dir = tf.keras.utils.get_file('flower_photos',origin=dataset_url,cache_dir='.',untar=True)
```

```python
import pathlib
data_dir = pathlib.Path(data_dir)
image_count = len(list(data_dir.glob('*/*.jpg')))
print(image_count)
```
